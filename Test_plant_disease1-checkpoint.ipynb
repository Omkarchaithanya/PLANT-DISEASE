## Testing Model

```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2

# Load the validation set to get class names
validation_set = tf.keras.utils.image_dataset_from_directory(
    'valid',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(128, 128),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
class_name = validation_set.class_names
print("Class Names:", class_name)

# Load the pre-trained model
model = tf.keras.models.load_model('Trained_model.keras')

# Define the image path
image_path ='C:/Users/CSP/archive/test/test/AppleScab1.JPG'

# Check if the file exists at the specified path
if not os.path.isfile(image_path):
    print(f"Error: File not found at {image_path}")
    # List the contents of the directory to verify the correct path
    directory = os.path.dirname(image_path)
    if os.path.isdir(directory):
        print(f"Contents of the directory '{directory}':")
        for file_name in os.listdir(directory):
            print(file_name)
else:
    # Reading the image using OpenCV
    img = cv2.imread(image_path)
    
    if img is None:
        print(f"Error: Unable to load image at {image_path}")
    else:
        # Convert BGR to RGB
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        # Display the original image
        plt.imshow(img_rgb)
        plt.title('Test Image')
        plt.xticks([])
        plt.yticks([])
        plt.show()
        
        # Preprocess the image for the model
        image = tf.keras.preprocessing.image.load_img(image_path, target_size=(128, 128))
        input_arr = tf.keras.preprocessing.image.img_to_array(image)
        input_arr = np.array([input_arr])  # Convert single image to a batch
        
        # Make predictions
        predictions = model.predict(input_arr)
        print("Predictions:", predictions)
        
        # Get the index of the highest prediction
        result_index = np.argmax(predictions)
        print("Predicted class index:", result_index)
        
        # Display the disease prediction
        model_prediction = class_name[result_index]
        plt.imshow(img_rgb)
        plt.title(f"Disease Name: {model_prediction}")
        plt.xticks([])
        plt.yticks([])
        plt.show()
```
